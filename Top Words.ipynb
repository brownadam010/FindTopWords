```python
#program to take the sense and sensibility book and list top words outside of the stopwords file
```

```python
bookFile = open("Sense_and_Sensibility.txt", "r")

stopWords = open("stopwords.txt", "r")

stop = stopWords.read().split("\n")

book = bookFile.read().split(" ")

bookFile.close()

histoWords = {}

badwords = {}


for i in book:
    similarWord = i.lower()
    if not similarWord.isalpha():
        continue
    if similarWord in stop:
        continue
    if similarWord in histoWords:
        histoWords[similarWord] = histoWords[similarWord] + 1
    else:
        histoWords[similarWord] = 1
        
sortedVal = sorted(histoWords.values(), reverse = True)

sortedHisto = {}
        
for i in sortedVal:
    for a in histoWords.keys():
        if histoWords[a] == i:
            sortedHisto[a] = histoWords[a]
            break

stars = "*"
            
for c in sortedHisto:
    print(c + " (" + str(sortedHisto[c]) + ") " + stars * round(float((sortedHisto[c]/10))))
      
        
    

    
```
